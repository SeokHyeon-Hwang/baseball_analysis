## 변수 설명
<pre>
* G : 게임수
* W : 승수
* T : 무승부(tie)
* D : 패배(Depeat)
* WR : 승률
* PW : 기대승률(피타고라스 승률?)
* R/G : 게임당 득점
* RS : 총 득점
* RA : 총 실점
* BA : 타율
* OBP : 출루율
* SLG : 장타율
* OOBP : 상대의 출루율
* OSLG : 상대의 장타율
</pre>

### RD(Run Difference) 변수 추가하기

```
### RS(총 득점)에 대한 히스토그램 그려보기
sns.distplot(df['RS'])
```

```
### RA(총 실점)에 대한 히스토그램 그려보기
sns.distplot(df['RA'])
```

```
### OBP(출루율)에 대한 히스토그램 그려보기
sns.distplot(df['OBP'])
```

```
### SLG(장타율)에 대한 히스토그램 그려보기
sns.distplot(df['SLG'])
```

```
### 승수 당 승점수에 대한 산점도 그래프
plt.figure(figsize=(7,4))
plt.scatter(x=df['RS'], y=df['W'], c='red')
plt.xlabel('Wins')
plt.ylabel('Run Scored')
```

```
### RD당 Wins 산점도 그래프 + 직선 그래프
x = np.array(df['RD'])
y = np.array(df['W'])

#### Deriving slope, intercept values
slope, intercept = np.polyfit(x, y, 1)
abline_values = [slope*i +intercept for i in x]

#### Plotting the figure
plt.figure(figsize=(7,5))
plt.scatter(x,y)
plt.plot(x, abline_values, 'b')
plt.title('slope = %s' % (slope))
plt.xlabel('Run Difference')
plt.ylabel('Wins')
plt.show()
```

```
### RD와 W의 상관계수 찍어보기
print(np.corrcoef(x, y))
```

## 모델 만들기
2015~2017년 데이터로 모델을 만들고 2018년 데이터로 모델 성능을 평가해 볼 것이다.

```
#### 1. RS 예측 모델 만들기
##### 데이터프레임에서 변수 추출
x = df.loc[df['year'] <= 2017][['OBP', 'SLG', 'BA']]
y = df.loc[df['year'] <= 2017][['RS']]
```

```
from sklearn.linear_model import LinearRegression
```

```
##### 모델 객체 호출하기
RS_model1 = LinearRegression()
```

```
##### 모델 피팅
RS_model1.fit(x, y)
```

```
##### 모델 절편 및 계수
print(RS_model1.intercept_)
print(RS_model1.coef_)
```

##### => RS모델은
##### RS = -1064 + 2816*(OBP) + 1511*(SLG) + 624*(BA)

##### 다중 공선선성을 고려하여 BA를 빼자.

```
##### 1. RS 예측 모델 2
###### 데이터프레임에서 변수 추출
x = df.loc[df['year']<=2017][['OBP', 'SLG']]
y = df.loc[df['year']<=2017][['RS']]
```

### 모델 평가
2017년도 데이터 넣어서 모델 평가

```
df2017= df.loc[df['year']==2017][['team','OBP', 'SLG', 'OOBP', 'OSLG', 'RS', 'RA', 'W']]
df2017
```

```
# 예측 모델에 넣기 위해 변수를 묶기 
df2017_for_RS = df2017[['OBP', 'SLG']]
df2017_for_RA = df2017[['OOBP', 'OSLG']]
```

```
# 모델로 2018년도 예측하기
df2018_RS = RS_model2.predict(df2017_for_RS)
df2018_RA = RA_model.predict(df2017_for_RA)
df2018_RD = df2018_RS - df2018_RA
df2018_W = W_model.predict(df2018_RD)
```

```
df2018_W.shape
```

```
type(df2018_W)
```

```
# shape 바꾸기 & 값들 보기좋게 소수점 버리기
import math
```

```
df2018_W_reshape = np.reshape(df2018_W, newshape=(10,))

df2018_Wr=[]
for i in df2018_W_reshape:
  dd = math.trunc(i)
  df2018_Wr.append(dd)
  
df2018_Wr
```

```
df2018_RS_reshape = np.reshape(df2018_RS, newshape=(10,))

df2018_RSr=[]
for i in df2018_RS_reshape:
  dd = math.trunc(i)
  df2018_RSr.append(dd)

df2018_RA_reshape = np.reshape(df2018_RA, newshape=(10,))

df2018_RAr=[]
for i in df2018_RA_reshape:
  dd = math.trunc(i)
  df2018_RAr.append(dd)

df2018_RD_reshape = np.reshape(df2018_RD, newshape=(10,))

df2018_RDr=[]
for i in df2018_RD_reshape:
  dd = math.trunc(i)
  df2018_RDr.append(dd)
  
print(df2018_RSr)
print(df2018_RAr)
print(df2018_RDr)
```

```
# 비교를 위한 데이터프레임 정리
## 2017년 데이터를 통해 뽑아낸 예측 데이터로 데이터프레임 만들기
## 추후에 이 데이터프레임을 2018년 데이터에 붙일 것.
d = {'team': df2017.team, 'PRS':df2018_RSr, 'PRA':df2018_RAr, 'PRD':df2018_RDr ,'PW':df2018_Wr}
df2018P = pd.DataFrame(data=d )
df2018P
```

```
# 2018년 데이터만 추출
df2018 = df.loc[df['year']==2018][['team', 'RS', 'RA', 'RD', 'W' ]]
df2018
```

```
# 팀 컬럼으로 2018예측 데이터프레임과 2018 데이터프레임 합치기
df2018S=df2018.set_index('team').join(df2018P.set_index('team'))
```

```
# 예측 RD 구하기
df2018S['PRD']=df2018S['PRS']-df2018S['PRA']
```

```
# 컬럼 정렬
cols = ['RS', 'PRS', 'RA', 'PRA', 'RD', 'PRD', 'W', 'PW']
df2018S = df2018S[cols]
```

```
# 2018예측 승수(PW:predicted Wins)로 정렬하여 출력하기
df2018S.sort_values(by='PW', ascending=False)
```
